# Tour de spark

```python
from pyspark.sql import SparkSession
APP_NAME = "Pyspark Tutorial"
```

Sans connection Ã  un master existant

```python
spark = SparkSession.builder\
    .appName(APP_NAME)\
    .getOrCreate()
```

Avec master local

```python
spark = SparkSession.builder\
    .appName(APP_NAME)\
    .master("spark://localhost:4040")\
    .getOrCreate()
```

```python
df = spark.read.csv('Chocolate Sales.csv',header=True)
df.head(3)
```

```python
df.show(5,0)
```

Compter le nombre de commerciaux

```python
df.select('Sales Person').distinct().count()
```

```python
spark.stop()
```
