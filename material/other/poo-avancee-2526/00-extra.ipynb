# Informations supplémentaires

Les jupyter notebooks sont des fichiers qui combinent du code exécutable et du texte formaté en markdown. Son extension est `.ipynb`.

**I**nteractive **PY**thon **N**ote**B**ook.

Pour l'utiliser avec VSCode, il faut installer l'extension [Jupyter](https://marketplace.visualstudio.com/items?itemName=ms-toolsai.jupyter) de Microsoft.

## Markdown

Le markdown est un langage comme le HTML mais en plus simple. D'ailleurs, markdown est un jeu de mots sur HTML car on qualifie HTML en anglais de "Markup language".

Exemples de syntaxe markdown courante :

`# Titre de niveau 1`

`## Titre de niveau 2`

`### Titre de niveau 3`

`**Texte en gras**` -> **Texte en gras**

`*Texte en italique*`

`- Liste à puces` (équivalet de ul en HTML - unordered list)

`1. Liste numérotée` (équivalet de ol en HTML - ordered list)

`[Texte du lien](URL)`

`![Texte alternatif](URL de l'image)`

`> Citation`
