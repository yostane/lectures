# POO en Java

En Java, il faut déclarer les propriété en `private` pour les rendre privées. Ensuite, les rendre visibles via des getters et setters publics. (En java, le code POO est plus verbeux qu'en Python).

`self` est remplacé par `this` (qui est implicite) en Java.

```python
class Car {
  private String name;
  private String brand;
  private int releaseYear;
  // Définition du constructeur
  public Car(String name, String brand, int releaseYear){
    this.name = name;
    this.brand = brand;
    this.releaseYear = releaseYear;
  }

  // Définition d'un getter/setter pour name
  public String getName(){
    return this.name;
  }
  public void setName(String name){
    this.name = name;
  }

  public String getBreand(){
    return this.brand;
  }

  public void setBrand(String brand){
    this.brand = brand;
  }

  public int getReleaseYear(){
    return this.releaseYear;
  }

  public void setReleaseYear(int releaseYear){
    this.releaseYear;
  }
}
```
