# POO en Java

En Java, il faut déclarer les propriété en `private` pour les rendre privées. Ensuite, les rendre visibles via des getters et setters publics. (En java, le code POO est plus verbeux qu'en Python).

`self` est remplacé par `this` (qui est implicite) en Java.

```java
class Car {
  private String name;
  private String brand;
  private int releaseYear;
  // Définition du constructeur
  public Car(String name, String brand, int releaseYear){
    this.name = name;
    this.brand = brand;
    this.releaseYear = releaseYear;
  }

  public void start(){
    System.out.println("Vroom Vroom");
  }

  // Définition d'un getter/setter pour name
  public String getName(){
    return this.name;
  }
  public void setName(String name){
    this.name = name;
  }

  public String getBreand(){
    return this.brand;
  }

  public void setBrand(String brand){
    this.brand = brand;
  }

  public int getReleaseYear(){
    return this.releaseYear;
  }

  public void setReleaseYear(int releaseYear){
    this.releaseYear = releaseYear;
  }
}
```

```java
Car car = new Car("Toyota", "Corolla", 2001);
car.start();
System.out.println(car.getName());
```

## Interfaces

Java ne propose pas d'héritage multiple de classes, mais permet d'implémenter plusieurs interfaces.

Une interface est une sorte de contrat qui définit un ensemble de méthodes qu'une classe doit implémenter. Une interface ne peut contenir que des déclarations de méthodes (sans implémentation) et des constantes.

Une classe qui implémente une interface doit fournir des implémentations concrètes pour toutes les méthodes déclarées dans l'interface.

```java
class Company { }
interface Building {
  public void openDoor();
}

class PrivateSchool extends Company implements Building {
  @Override
  public void openDoor() {
    System.out.println("School is open");
  }
}

PrivateSchool privateSchool = new PrivateSchool();
privateSchool.openDoor();
```
