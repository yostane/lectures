# Analyse du dataset titanic

```python
import pandas as pd

df = pd.read_csv("titanic.csv")
df.head() # Affiche les premières lignes
```

- Calculer le nombre de survivants et de non survivants. (`value_counts()` compte les occurrences de chaque valeur dans une colonne)

```python
df["Survived"].value_counts()
```

- Calculer le nombre de femmes et d'hommes à bord du Titanic.

```python
df["Sex"].value_counts()
```

- Pourcentage de survivants.

```python
survived_p = 100 * (df["Survived"].sum() / df["Survived"].count())
round(survived_p, 2)
```

```python
import math
math.ceil(survived_p)
```

- Nombre de survivants par classe de billet.

```python
df[["Pclass", "Survived"]].head()
```

```python
df[["Pclass", "Survived"]].groupby("Pclass").sum()
```

```python
survivors = df[df["Survived"] == 1]
survivors.groupby("Pclass").sum()
```

- Compter le nombre de femmes par classe de billet.

```python
females = df[df["Sex"] == "female"]
females[["Sex", "Pclass"]].value_counts()
```
