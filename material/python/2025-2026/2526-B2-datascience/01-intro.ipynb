```python
import pandas as pd # pd est l'alias de pandas

data = {
    "Année": [2024, 2025, 2026, 2027],
    "Diplôme": ["BAC", "B1", "B2", "B3"]
}

df = pd.DataFrame(data)
df
```

```python
df["Année"].max()
```

describe permet de donner des statistiques descriptives sur un DataFrame

```python
df.describe()
```

```python
moyennes = {
    "Année": [2024, 2025, 2026, 2027],
    "Diplôme": ["BAC", "B1", "B2", "B3"],
    "Moyenne": [2, 19.4, 10, 15.8]
}

df_moyennes = pd.DataFrame(moyennes)
df_moyennes
```

```python
import matplotlib.pyplot as plt

plt.plot(df_moyennes["Année"], df_moyennes["Moyenne"])
plt.show()
```
