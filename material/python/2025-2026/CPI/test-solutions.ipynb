## Part 3: Coding questions

1. Program that generates a list of 15 random integers between 42 and 60 and prints requested values (without using sum/max/min built-ins).

Solution (complete script):

```py
import random

nums = []
for _ in range(15):
    nums.append(random.randint(42, 60))

# print list
print("The content of the list is:", nums)

# compute sum and average without sum()
total = 0
for v in nums:
    total += v
avg = total / len(nums)
print(f"Sum: {total}, Average: {avg}")

# compute min and max without min()/max()
min_val = nums[0]
max_val = nums[0]
for v in nums:
    if v < min_val:
        min_val = v
    if v > max_val:
        max_val = v
print(f"Max: {max_val}, Min: {min_val}")
```

2. Define play_game function

Solution:

```py
def play_game(target, attempts):
    for _ in range(attempts):
        guess = int(input("Enter your guess: "))
        if guess < target:
            print("Too low!")
        elif guess > target:
            print("Too high!")
        else:
            print("Correct! You win!")
            return
    print(f"Sorry, you've used all attempts. The correct number was {target}.")
```

3. Fibonacci function

Solution (iterative for efficiency):

```py
def fib(n):
    if n == 0:
        return 0
    if n == 1:
        return 1
    a = 0
    b = 1
    c = None
    for _ in range(2, n + 1):
        c = a + b
        a = b
        b = c
    return c
```

Example:
- fib(6) -> 8
- fib(10) -> 55
