- Create a program that asks the user his favorite movies. The user can input as many movies as he wants, one by one. When the user inputs the letter `"q"` (for "quit"), the program stops asking for more movies. Store the movies in a list, then save the list to a file in JSON format. Finally, read back the list from the file and print the number of movies.

````python
import json

movies = []
movie = input("Enter a movie name or 'q' to quit")
while movie != "q":
  movies.append(movie)
  movie = input("Enter a movie name or 'q' to quit")

with open("movies.json", "w") as file:
    file.write(json.dumps(movies))
    
with open("movies.json") as file:
  loaded_movies = json.load(file)
  print("Movies loaded from the file", loaded_movies)
````

- Create a program that plays guess the number game while saving the current progress. The program randomly selects a number between 1 and 10, and the user has to guess it after 5 attempts. After each guess, the program tells the user if the guess is too low, too high, or correct. The program saves the user's guesses to a file in JSON format after each guess. If the user exits the program and restarts it later, the program loads the previous guesses from the file and continues from where the user left off. If the user wins or loses, the saved game is deleted.

Game without save and load functionality:

````python
import json
import random

def play_game():
  r = random.randint(1, 10)
  for i in range(5):
    user_proposition = int(input(f"Guess a number (attempt {i})"))
    if user_proposition == r:
      print("You won!")
      return
  print("you lost")
  
play_game()
````

Game with save functionality:

````python
import json
import random

def save_game(game):
  with open("game.json", "w") as file:
    file.write(json.dumps(game))

def play_game():
  r = random.randint(1, 10)
  game = {"number_to_guess": r, "user_propositions": []}
  save_game(game)
  for i in range(5):
    user_proposition = int(input(f"Guess a number (attempt {i})"))
    game["user_propositions"].append(user_proposition)
    save_game(game)
    if user_proposition == r:
      print("You won!")
      return
  print("you lost")

play_game()
````

Game with save and load functionality. If the user wins or loses, the saved game is deleted:

````python
import json
import random
from pathlib import Path

def delete_save():
  print("Deleting save")
  Path("game.json").unlink()

def save_game(game):
  with open("game.json", "w") as file:
    file.write(json.dumps(game))
    
def load_game():
  if not Path("game.json").exists():
    return None
  with open("game.json") as file:
    game = json.load(file)
    print("Loaded game", game)
    return game

def play_game():
  game = load_game()
  if game == None:
    r = random.randint(1, 10)
    game = {"number_to_guess": r, "user_propositions": []}
    save_game(game)
  r = game["number_to_guess"]
  start_attempt = len(game["user_propositions"])
  for i in range(start_attempt, 5):
    user_proposition = int(input(f"Guess a number (attempt {i})"))
    game["user_propositions"].append(user_proposition)
    save_game(game)
    if user_proposition == r:
      print("You won!")
      delete_save()
      return
  print("you lost")
  delete_save()

play_game()
````

  Load the JSON data from the file and print this information:

  - The full name of each film maker (first name + last name) and the number of movies they directed.
  - The average birth year of all film makers in the file.
  - A movie with the longest title. If there are multiple movies with the same length, print one of them.
  - The movies with the longest title. If there are multiple movies with the same length, print all of them.

````python
import json

with open("data.json") as file:
  movies_makers = json.load(file)
  print("The full name of each film maker (first name + last name) and the number of movies they directed.")
  sum = 0
  for movie_maker in movies_makers:
    print(movie_maker["first_name"], movie_maker["last_name"], ". Number of movies:", len(movie_maker["movies"]))
    sum += movie_maker["birth_year"]
  avg = sum // len(movies_makers)
  print("The average birth year of all film makers in the file.", avg)
  movie_with_longest_title = ""
  longest_title = 0
  for movie_maker in movies_makers:
    for movie in movie_maker["movies"]:
      if len(movie) > longest_title:
        longest_title = len(movie)
        movie_with_longest_title = movie
  movies_with_longest_titles = []
  print("Longest title length", longest_title)
  print("One movie among the ones with the longest title.", movie_with_longest_title)
  for movie_maker in movies_makers:
    for movie in movie_maker["movies"]:
      if len(movie) == longest_title:
        movies_with_longest_titles.append(movie)
  print("Movies with the longest titles.", movies_with_longest_titles)
````
