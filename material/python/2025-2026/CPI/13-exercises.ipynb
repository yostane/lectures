# Exercises

- Create a dictionary representing a book, with keys for title, author, year published, and genre. Print the dictionary. Save it to a file in JSON format, then read it back and print it again.

- Create a program that simulates an ordering machine in a restaurant. The program asks the user to **input** the name of his entrée, then, the name of his main course, then the name of his dessert. Store the order in a dictionary with keys `"entrée"`, `"main course"`, and `"dessert"`. Finally, print the order summary to the user and save it to a file.

- Create a program that simulates an improved ordering machine in a restaurant (compared to the previous exercise). The program asks the user to **input** the name of his entrée, then, the name of his main course, then the name of his dessert. Store the order in a dictionary with keys `"entrée"`, `"main course"`, and `"dessert"`. Next (the improevement), the program asks the user if he wants to add a drink to his order (yes/no). If the user answers "yes", the program asks for the name of the drink and adds it to the order dictionary with the key `"drink"`. Next (another improvement), the program shows the summary of the order and asks the user to confirme (yes/no). If the user answers "no", the program allows the user to modify his order by re-inputting the entrée, main course, dessert, and optionally drink. Finally when the user confirms the order, the program saves it to a file.

- Create a program that asks the user his favorite movies. The user can input as many movies as he wants, one by one. When the user inputs the letter `"q"` (for "quit"), the program stops asking for more movies. Store the movies in a list, then save the list to a file in JSON format. Finally, read back the list from the file and print the number of movies.

- Let's suppose that we already have a file named `data.json` with the following content (you can create it manually and copy-paste the content):
