```python
import pandas as pd
import matplotlib.pyplot as plt

air_quality = pd.read_csv("air_quality_no20.csv", index_col=0, parse_dates=True)
air_quality.head()
```

## Différents types de graphiques

```python
air_quality.plot()
plt.show()
```

```python
air_quality["station_paris"].plot()
plt.show()
```

```python
air_quality[air_quality["station_paris"] > 20].plot()
plt.show()
```

```python
air_quality.plot.scatter(x="station_london", y="station_paris", alpha=0.5)
plt.show()
```

```python
air_quality.plot.box()
plt.show()
```

```python
# subplots permet de faire un graphique par colonne
# area permet de tracer un graphique rempli
axs = air_quality.plot.area(figsize=(12, 4), subplots=True)
plt.show()
```

```python
# subplot permet de créer un graphique personnalisable
fig, axs = plt.subplots(figsize=(12, 4))
air_quality.plot.area(ax=axs)
axs.set_ylabel("NO$_2$ concentration")
fig.savefig("no2_concentrations.png")
plt.show()
```

```python
fig, axs = plt.subplots(2)
fig.suptitle('Vertically stacked subplots')
air_quality.plot.area(ax=axs[0])
axs[0].set_ylabel("NO$_2$ concentration")
air_quality.plot.box(ax=axs[1])
plt.show()
```
